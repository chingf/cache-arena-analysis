```python
import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

theta = np.linspace(0, 2 * np.pi, 201)
x = np.cos(theta)
y = np.sin(theta)

ax.plot(x, y, 0, linewidth=2, label="Place")
ax.scatter(
    [0], [0], [0.25], linewidth=2,
    label="Episode", color="red")
ax.legend()
ax.set_xticklabels([])
ax.set_yticklabels([])
ax.set_zticklabels([])

plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

theta = np.linspace(0, 2 * np.pi, 201)
x = np.cos(theta)
y = np.sin(theta)

num_cs = 16
theta = np.linspace(0, 2 * np.pi, num_cs)
newx = np.cos(theta)
newy = np.sin(theta)
newx = newx + np.random.normal(0, 0.75, num_cs)
newy = newy + np.random.normal(0, 0.75, num_cs)
newz = np.random.normal(0, 0.75, num_cs)
newx[-1] = newx[0]
newy[-1] = newy[0]
newz[-1] = newz[0]

ax.plot(x, y, 0, linewidth=2, label="Place")
ax.plot(
    newx, newy, newz, linewidth=2, label="Episode",
    color="red", alpha=0.7
    )
ax.legend()
ax.set_xticklabels([])
ax.set_yticklabels([])
ax.set_zticklabels([])

plt.show()
```
